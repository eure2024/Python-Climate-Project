```python
df=pd.read_csv('us_emissions.csv')
```

```python
#filter
newdf=df[(df.Country=="USA") & (df.Year>1949)]
```

```python
newdf
```

```python
#line graph
df=pd.DataFrame(newdf)
plt.plot(df['Year'],df['Total'],color='red',marker='.',markerfacecolor='black')
plt.title('United States Carbon Emissions -WB',fontsize=14)
plt.xlabel('Year')
plt.ylabel('Carbon Emissions (mt)')
plt.text(1973+2, 4200, "A", ha='center')
plt.text(1979+3.5, 4200, "B", ha='center')
plt.text(2020, 4500, "C", ha='center')
plt.text(1900, 6200, "A: 1973 Oil Crisis\nB: 1979 Oil Crisis\nC: COVID-19 Lockdowns", fontsize=12, ha='left', va='top')
plt.show()
```

This graph illustrates the amount of carbon emissions produced by the U.S.A, measured in metric tons, ranging from 1950 to 2021. Points A, B, and C on the graph represent significant historical events contributing to rapid short-term declines in carbon emissions.

```python
df_grouped = df.groupby('Year').sum()[['Coal','Gas','Oil','Cement','Flaring','Other']]
df_grouped
```

```python
fig,ax = plt.subplots (1,figsize=(50,6))
x=np.arange(0,len(df_grouped.index))
plt.bar(x-0.3,df_grouped['Coal'],width=0.2,color='#1D2f6F')
plt.bar(x-0.1,df_grouped['Gas'],width=0.2,color='#8390FA')
plt.bar(x+0.1,df_grouped['Oil'],width=0.2,color='#6EAF46')
plt.bar(x+0.3,df_grouped['Cement'],width=0.2,color='#FAC748')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.ylabel('Total Emission (in Metric Ton)')
plt.xticks(x,df_grouped.index)
plt.xlim(-0.5,72)

ax.set_axisbelow(True)
ax.yaxis.grid(color='gray',linestyle='dashed',alpha=0.2)

plt.title('Breakdown Emission Per Year by Category - XY',loc='left')
plt.legend(['Coal','Gas','Oil','Cement'],loc='upper left',ncol=4)
plt.show()



#Click on graph to expand size
```

Unfortunately, this graph shows that coal and gas contributed to a majority of carbon emissions which suggests we should look for energy sources to replace coal and oil, such as wind and sun.

```python
#temperature scatter plot

plt.rcParams["figure.figsize"] = (12,6)

plt.scatter(temperature['Date'],temperature['Value'],color='orange')
plt.title('United States Average Temperatures - EU',fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temperatures (°F)', fontsize=14)
plt.text(1962.5,52.55, "1950-2020 Mean: 52.4°F", ha='center')
plt.text(1962.5, 53.75, "2000-2020 Mean: 53.5°F", ha='center')
plt.axhline(y=53.5, color= 'b', linestyle = ':')
plt.axhline(y=52.4, color= 'b', linestyle = ':')
plt.show()

```

This graph displays annual average temperatures in the U.S.A. from 1950 through 2020 and two lines that highlight the mean temperature from 1950 through 2020 and 2000 through 2020 to show the increase in average temperatures over the past 20 years.

```python
#total of propertydamage by year

propertydamage['episode_month'] = propertydamage['episode_month'].astype(int)

groupdf=propertydamage.groupby('episode_month',sort=True).sum()
groupdf

sortedDf=groupdf.sort_values('damage_property_in_billions', ascending=False)
df2 = sortedDf.head(5).plot(kind='pie', y='damage_property_in_billions', fontsize=14, autopct='%1.0f%%')
plt.title('Years with the Most Property Damage - EU',fontsize=14)
df2


```

This graph displays the top 5 years with the most property damage from 1950 through 2020 (measured in billions) as a result of severe weather events in the U.S.A. which are a result of increased CO2 levels and average annual temperatures.
